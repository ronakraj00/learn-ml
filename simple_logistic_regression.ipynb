```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([2,3,4,5,6,7,8])
y=np.array([0,0,0,1,1,1,1])
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
```

```python
dev_x=x-mean_x
dev_y=y-mean_y
```

```python
prod_dev=dev_x*dev_y
```

```python
sum_prod_dev=np.sum(prod_dev)
```

```python
sq_dev_x=dev_x**2
```

```python
m=sum_prod_dev/np.sum(sq_dev_x)
```

```python
b=mean_y-m*mean_x
```

```python
print(m,b)
```

```python
plt.scatter(x,y)
```

```python
sigmoid=1/(1+np.exp(-(m*x+b)))
```

```python
y_pred=np.where(sigmoid>0.5,1,0)
```

```python
plt.plot(x,y_pred)
```
